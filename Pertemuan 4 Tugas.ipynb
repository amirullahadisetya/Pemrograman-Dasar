```python
# Tugas 1

print("="*30)
print("SELAMAT DATANG DI KEDAI AGAM")
print("="*30)
print("Menu Makanan & Minuman:")
print("1. Whice Coffe    - Rp 8.000")
print("2. Teh Telur      - Rp 13.000")
print("3. Mie Aceh       - Rp 15.000")
print("4. Mie Bangladesh - Rp 20.000")
print("="*30)

# Input Pilihan Menu
pilihan_str = input("silakan masukan nomor pilihan Anda (1-4): ")

# Inisialisasi variabel
pilihan = int(pilihan_str) # Konversi input ke integer
nama_pesanan = ""
harga = 0

# Penentuan harga dan nama pesanan menggunakan if/elif/else
if pilihan == 1:
    nama_pesanan = "White Coffe"
    harga = 8000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 2:
    nama_pesanan = "Teh Telur"
    harga = 13000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 3:
    nama_pesanan = "Mie Aceh"
    harga = 15000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
elif pilihan == 4:
    nama_pesanan = "Mie Bangladesh"
    harga = 20000
    print(f"Anda memesan: {nama_pesanan} dengan harga Rp {harga}")
else:
    print("Maaf, pilihan menu tidak valid. Silakan ulang program.")
    # Kita tidak akan melanjutkan program jika pilihan tidak valid
    # Untuk kasus ini, kita biarkan harga tetap 0 sehingga program berhenti di bawah

# --- BAGIAN MODIFIKASI DISKON DIMULAI DI SINI ---
total_harga = harga # Simpan harga asli sebelum diskon
diskon_persen = 0
diskon_rupiah = 0

if total_harga > 0:
    # Cek kategori diskon berdasarkan total_harga
    if total_harga >= 50000:
        diskon_persen = 70
    elif total_harga >= 25000:
        diskon_persen = 50
    else: # total_harga < 25000
        diskon_persen = 30

    # Hitung diskon
    diskon_rupiah = total_harga * (diskon_persen / 100)
    harga_setelah_diskon = total_harga - diskon_rupiah

    print("-" * 30)
    print(f"Total Belanja Awal: Rp {total_harga:,}")
    print(f"Anda mendapatkan DISKON: {diskon_persen}%")
    print(f"Potongan Diskon: Rp {diskon_rupiah:,.0f}") # Menggunakan .0f agar diskon dibulatkan

    # Ganti variabel 'harga' dengan harga_setelah_diskon untuk perhitungan pembayaran
    harga = int(harga_setelah_diskon)
    print(f"TOTAL YANG HARUS DIBAYAR: Rp {harga:,}")
    print("-" * 30)
# --- BAGIAN MODIFIKASI DISKON SELESAI DI SINI ---

if harga > 0:
    # Proses Pembayaran
    uang_dibayar_str = input(f"Total yang harus dibayar: Rp {harga:,}. Masukkan jumlah uang Anda: Rp ")
    uang_dibayar = int(uang_dibayar_str)

    if uang_dibayar >= harga:
        kembalian = uang_dibayar - harga
        print("\n--- Struk Pembayaran ---")
        print(f"Pesanan Anda  : {nama_pesanan}")
        print(f"Total Harga   : Rp {total_harga:,}") # Tampilkan harga awal
        print(f"Diskon        : -Rp {diskon_rupiah:,.0f}")
        print(f"Total Dibayar : Rp {harga:,}") # Tampilkan harga setelah diskon
        print(f"Uang Pembayaran : Rp {uang_dibayar:,}")
        print(f"Kembalian     : Rp {kembalian:,}")
        print("------------------------")
        print("Terima kasih telah berbelanja!")
    else:
        kekurangan = harga - uang_dibayar
        print(f"Maaf, uang Anda kurang sebesar Rp {kekurangan:,}")

print("="*30)
```
